```python
from google.cloud import storage
client = storage.Client()
bucket = client.get_bucket('ml_project_hns')

def readCsv(input_fname: str, bucket_name:storage.bucket.Bucket)-> pd.DataFrame:
    blob = storage.blob.Blob(input_fname,bucket_name)
    content = blob.download_as_string()
    return(pd.read_csv(BytesIO(content)))
```

```python
# Read the datasets
df_kaggle_train = readCsv('application_train.csv', bucket)
df_kaggle_test = readCsv('application_test.csv', bucket)
```

```python
print("Shape of Train data:", df_kaggle_train.shape)
print("Shape of Test data:", df_kaggle_test.shape)
```

```python
# The difference in shape of the two dataframes is because for test dataset, we dont have the TARGET variable
print(df_kaggle_train['TARGET'].unique())
# Add target variable into test dataset with value that is not present in the train dataset TARGET variable
#  so that the two dataframes can be recreated
df_kaggle_test['TARGET'] = -1
# Join the test and train dataframes
df_kaggle = pd.concat([df_kaggle_train, df_kaggle_test], sort=False)
df_kaggle.shape
```
