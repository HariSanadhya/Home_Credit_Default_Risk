```python
def readCsv(input_fname: str, bucket_name:storage.bucket.Bucket)-> pd.DataFrame:
    blob = storage.blob.Blob(input_fname,bucket_name)
    content = blob.download_as_string()
    return(pd.read_csv(BytesIO(content)))
```

### df_bureau_bal data analysis - convert it to more usable form

```python
df_bureau_bal.head(10)
```

```python
# Get the count of each distinct status of every SK_ID_BUREAU
df_bureau_bal_modified = df_bureau_bal.groupby(['SK_ID_BUREAU', 'STATUS']).count().reset_index()

# Restructure the dataframe to have single record for each SK_ID_BUREAU with distinct status count as columns
df_bureau_bal_modified = df_bureau_bal_modified.pivot(index='SK_ID_BUREAU', columns='STATUS', 
            values='MONTHS_BALANCE')

# Replace all null values by zero
df_bureau_bal_modified.fillna(0, inplace=True)

# Compute the age of the SK_ID_BUREAU in months
df_bureau_bal_modified['AGE']= df_bureau_bal_modified.sum(axis=1)

df_bureau_bal_modified.reset_index(inplace=True)

# If the SK_ID_BUREAU is having closed status (column C), then treat it as flag(0,1 value)
df_bureau_bal_modified['C'] = df_bureau_bal_modified['C'].apply(lambda x:1 if x>0 else 0)

df_bureau_bal_modified.head()
```

# Deep Feature synthesis
Below is the list of all the possible afeature types that can be geenrated using deep feature systhesis. Some of them are aggregation type while rest are transformation. Not all the types of features are required to be generated. So we picked some of the commonly used features names and generated corresponding deep features.

```python
pd.DataFrame(ft.list_primitives()).head(70)
```

```python
# Feature names selected to perform deep feature geenration using featuretools
default_agg_primitives =  ["sum", "max", "mean", "count"]
default_trans_primitives =  []
```

## EntitySet Creation

```python
# Create entityset
es = ft.EntitySet(id="home_credit")
```

## Entities Addition

```python
# Entities with a unique index
es = es.entity_from_dataframe(entity_id="application",
                                  dataframe=df_kaggle,
                                  index="SK_ID_CURR",
                                  variable_types={"TARGET": ft.variable_types.Categorical})

es = es.entity_from_dataframe(entity_id="bureau",
                                  dataframe=df_bureau_onehot,
                                  index= "SK_ID_BUREAU")

es = es.entity_from_dataframe(entity_id="prev_app",
                                  dataframe=df_previous_onehot,
                                  index = 'SK_ID_PREV')

es = es.entity_from_dataframe(entity_id="pos_cash",
                                  dataframe=df_cash_curr,
                                  index = 'SK_ID_PREV')

es = es.entity_from_dataframe(entity_id="payment_app",
                                  dataframe=df_payment_apps,
                                  index = 'SK_ID_PREV')

es = es.entity_from_dataframe(entity_id="credit_app",
                                  dataframe=df_credit_curr,
                                  index = 'SK_ID_PREV')
```

## Relationship between entities
![title](images/home_credit_ER.png)
<b>Based of the diagram above (provided by Home Credit on Kaggle) which shows how the tables are related, define relationships in featuretools.</b>

```python
# Add Relationship

# Relationship between app and bureau
r_app_bureau = ft.Relationship(es['application']['SK_ID_CURR'], es['bureau']['SK_ID_CURR'])

# Relationship between current app and previous apps
r_app_previous = ft.Relationship(es['application']['SK_ID_CURR'], es['prev_app']['SK_ID_CURR'])

# Relationships between previous apps and cash, installments, and credit
r_previous_cash = ft.Relationship(es['prev_app']['SK_ID_PREV'], es['pos_cash']['SK_ID_PREV'])
r_previous_installments = ft.Relationship(es['prev_app']['SK_ID_PREV'], es['payment_app']['SK_ID_PREV'])
r_previous_credit = ft.Relationship(es['prev_app']['SK_ID_PREV'], es['credit_app']['SK_ID_PREV'])

# Add in the defined relationships into the entity-set
es = es.add_relationships([r_app_bureau, r_app_previous, r_previous_cash, 
                          r_previous_installments, r_previous_credit])
```

```python
# Print entitySet to view the summary of the elements in it                  
print(es)
```

### Deep Feature Synthesis

```python
print('Deep Feature Synthesis in Progress')


# DFS for application features 
feature_matrix, feature_names = ft.dfs(entityset = es, target_entity = 'application',
                       trans_primitives = default_trans_primitives,
                       agg_primitives=default_agg_primitives, 
                       max_depth = 2, features_only=False, verbose = True)
# train_feature_matrix, train_feature_names = ft.dfs(entityset = es, target_entity = 'application_train',
#                        trans_primitives = default_trans_primitives,
#                        agg_primitives=default_agg_primitives, 
#                        max_depth = 2, features_only=False, verbose = True)
# test_feature_matrix, test_feature_names = ft.dfs(entityset = es, target_entity = 'application_test',
#                        trans_primitives = default_trans_primitives,
#                        agg_primitives=default_agg_primitives, 
#                        max_depth = 2, features_only=False, verbose = True)
                       
# Reset the index to make SK_ID_CURR a column again
feature_matrix = feature_matrix.reset_index()
# train_feature_matrix = train_feature_matrix.reset_index()
# test_feature_matrix = test_feature_matrix.reset_index()
```
